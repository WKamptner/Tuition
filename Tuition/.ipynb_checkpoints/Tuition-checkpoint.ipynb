```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# US Average College Tuition

### This dataset was provided by the tidytuesday datascience group on github and provides the average cost of college tuition for each state between the years of 2004 to 2016.

### The goal of this exercise is to produce meaningful visuals 

```python
df = pd.read_excel('us_avg_tuition.xlsx')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# Which state has the highest average tuition?
```

```python
n = len(df.columns)
```

```python
# Taking the sum of all columns ignoring the State
df['Sum']=df.iloc[:,1:n-1].sum(axis=1)
```

```python
# Taking the average by diving the Sum by the number of tuition fields
df['Avg']=df['Sum']/n-1
```

```python
# Next we will take a look at the average cost of tuition and look at which states are the most expensive
# and which are the most affordable
```

```python
df[['State','Avg']].sort_values(by='Avg')
```

```python
# Vermont is our most expensive state for tuition, on average, while Wyoming is our cheapest
```

```python
# Let's take a look at Vermont in particular and do a trend line
# let's first reshape our dataframe so it's a bit easier to look at
reshaped_df = pd.melt(df, id_vars=['State'], var_name='Year')
```

```python
# This looks a lot easier to read, next we'll plot this on a line graph to see what the trend looks like.
# But first, let's drop the SUM and AVG
reshaped_df[reshaped_df['State'] == 'Vermont']
```

```python
values = ['Sum', 'Avg']
```

```python
reshaped_df = reshaped_df[reshaped_df.Year.isin(values) == False]
```

```python
# It's sad to see it gone but we must do what we must do.  Once again, this is an exercise in simple data analytics
# and it's always good to practice python functionality
```

```python
reshaped_df[reshaped_df['State'] == 'Vermont']
```

```python
# set the size of our plot
sns.set(rc = {'figure.figsize':(15,8)})
```

```python
# plot Vermont tuition over the years
sns.lineplot(x = "Year", y="value", data = reshaped_df[reshaped_df['State'] == 'Vermont'])
```

```python
Years = reshaped_df['Year'].unique().tolist()
```

```python
Vermont_ln=reshaped_df[reshaped_df['State'] == 'Vermont']
Wyoming_ln=reshaped_df[reshaped_df['State'] == 'Wyoming']
Virginia_ln=reshaped_df[reshaped_df['State'] == 'Virginia']
```

```python
sns.lineplot(x = "Year", y="value", data = Vermont_ln, color = "blue")
sns.lineplot(x = "Year", y="value", data = Wyoming_ln, color = "red")
sns.lineplot(x = "Year", y="value", data = Virginia_ln, color = "green")
```
