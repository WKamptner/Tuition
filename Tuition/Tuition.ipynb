### This dataset was provided by the tidytuesday datascience group on github and provides the average cost of college tuition for each state between the years of 2004 to 2016.

```python
# Which state has the highest average tuition?
```

```python
n = len(df.columns)
```

```python
# Taking the sum of all columns ignoring the State
df['Sum']=df.iloc[:,1:n-1].sum(axis=1)
```

```python
# Taking the average by diving the Sum by the number of tuition fields
df['Avg']=df['Sum']/n-1
```

```python
# Next we will take a look at the average cost of tuition and look at which states are the most expensive
# and which are the most affordable
```

```python
df[['State','Avg']].sort_values(by='Avg')
```
